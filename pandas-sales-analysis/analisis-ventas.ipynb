# Preguntas

## 1️⃣ Leer el archivo
- Importa Pandas y carga el archivo `Ventas.csv` en un DataFrame llamado `df`.
- Muestra las primeras 7 filas.

```python
#Importar librerías necesarias
import pandas as pd
```

```python
#Cargar el archivo CSV en un DataFrame
df = pd.read_csv('data/raw/Ventas.csv')

#Mostrar las primeras 7 filas del DataFrame
primeras_7_filas = df.head(7)
print(primeras_7_filas)
```

## 2️⃣ Inspección

- Cuantas filas y columnas tiene el Dataframe.

```python
#Calcular y mostrar numero e filas con shape
numero_filas = df.shape[0]
print(f'Número de filas: {numero_filas}')

#Calcular y mostrar numero de columnas con shape
numero_columnas = df.shape[1]
print(f'Número de columnas: {numero_columnas}')
```

## 3️⃣ Selección de columnas

- Muestra solo las columnas Cliente y Producto.

```python
#Crear nueva df con solo las columnas Cliente y Producto
df_cliente_producto = df[['Cliente', 'Producto']]
print(df_cliente_producto)
```

## 4️⃣ Filtrado de datos

- Muestra todas las ventas que se hicieron en la zona "Norte".

```python
#Calcular y mostrar solo las ventas de la zona Norte
ventas_zona_norte = df[df['Zona'] == 'Norte']
print(ventas_zona_norte)
```

## 5️⃣ Operaciones Básicas

- Crea una nueva columna Total con el cálculo de (Cantidad * Precio).
- Muestra el DataFrame actualizado ( 8 primeras filas)

```python
#Crear nueva columna Total multiplicando Cantidad por Precio
df['Total'] = df['Cantidad'] * df['Precio']

#Guardar en variable las primeras 7 filas del DataFrame con la nueva columna Total
primeras_7_filas_con_total = df.head(7)

#Mostrar las primeras 7 filas del DataFrame con la nueva columna Total
print(primeras_7_filas_con_total)
```

## 6️⃣ Análisis

- ¿Cuál fué el total de ventas del día?
- ¿Cuál es la venta por cada zona?

```python
#Calcular y mostrar total de ventas del día
total_ventas_dia = df['Total'].sum()

print(f'Total de ventas del día: {total_ventas_dia}')
print('-' * 40)

#Calcular y mostrar total de ventas por zona
total_ventas_por_zona = df.groupby('Zona')['Total'].sum()

print(f'Total de ventas por zona:\n{total_ventas_por_zona}')
```

## 7️⃣ Guardar Resultados

- Generar un archivo "resumen.csv" que contenga:

    - Suma de Ventas por Zona (Descripcion y Total)
    - Suma de Ventas por producto (Descripcion y Total)
    - Cantidad de registros (transacciones) : xxxxx
    - Suma de Ventas: yyyyy

- El archivo debiera tener 3 columnas:
    - Descripcion 
    - Valor
    - Tipo

- Ejemplo:


| Descripción | Valor  | Tipo      |
|------------|-----------|-----------|
| Centro     | 711,993   | Zonas     |
...
| Sur        | 624,513   | Zonas     |
| Banana     | 429,714   | Productos |
| Cereza     | 107,314   | Productos |
...
| Piña       | 335,352   | Productos |
| Uva        | 352,107   | Productos |
| venta      | 2,702,306 | Resumen   |
| registros  | 1,000     | Resumen   |



Ayuda: Algunas maneras de renombrar Las columnas de un dataframe:

- df = df.rename(columns={'nombre_antiguo': 'nombre_nuevo'})
- df.rename(columns={"A": "Columna1", "B": "Columna2"}, inplace=True)
- df.columns = ['nuevo_nombre1', 'nuevo_nombre2', ...]

```python
#Crear variables resumen_zonas, resumen_productos y resumen_general con los totales de ventas por zona, producto y un resumen general respectivamente. Luego concatenar estas variables en una sola variable resumen y formatear la columna Valor para mostrar los valores con formato de moneda. Finalmente, guardar el resultado en un archivo CSV llamado resumen.csv y mostrar el resultado.
resumen_zonas = df.groupby('Zona')['Total'].sum().reset_index()
resumen_zonas.columns = ['Descripcion', 'Valor']
resumen_zonas['Tipo'] = 'Zonas'

resumen_productos = df.groupby('Producto')['Total'].sum().reset_index()
resumen_productos.columns = ['Descripcion', 'Valor']
resumen_productos['Tipo'] = 'Productos'

resumen_general = pd.DataFrame({
    'Descripcion': ['venta', 'registros'],
    'Valor': [df['Total'].sum(), len(df)],
    'Tipo': ['Resumen', 'Resumen']
})

# Concatenar los resúmenes y formatear la columna Valor
# Se usó IA paara ayudar a formatear la columna valor.

resumen = pd.concat([resumen_zonas, resumen_productos, resumen_general], ignore_index=True)
resumen['Valor'] = resumen.apply(
    lambda fila: f"${fila['Valor']:,.0f}".replace(',', '.') if fila['Descripcion'] != 'registros' else f"{fila['Valor']:,.0f}".replace(',', '.'),
    axis=1
)

# Guardar el resultado en un archivo CSV llamado resumen.csv
resumen.to_csv('data/output/resumen.csv', index=False)

# Mostrar el resultado
print(resumen)
```
